```python
class Node:
    def __init__(self, key, data):
        self.key = key
        self.data = data
        self.height = 1
        self.left = None
        self.right = None

class ArbolAvl:
    #insertar  un nodo en el arbol AVL
    def insertar(self, root, key, data):
        if not root:
            return Node(key, data)
        elif key < root.key:
            root.left = self.insertar(root.left, key, data)
        else:
            root.right = self.insertar(root.right, key, data)

        # Actualizar la altura del nodo actual
        root.height = 1 + max(self.altura(root.left), self.altura(root.right))

        # Obtener el factor de equilibrio del nodo para verificar el desequilibrio
        balance = self.balance(root)

        # Casos de desequilibrio y rotaciones
        #izquierda izquierda
        if balance > 1 and key < root.left.key:
            return self.rotate_right(root)
        #derecha derecha
        if balance < -1 and key > root.right.key:
            return self.rotate_left(root)
        #izquierda derecha
        if balance > 1 and key > root.left.key:
            root.left = self.rotate_left(root.left)
            return self.rotate_right(root)
        #derecha izquierda
        if balance < -1 and key < root.right.key:
            root.right = self.rotate_right(root.right)
            return self.rotate_left(root)

        return root
    
    def eliminar(self):
        pass

    def buscar(self):
        pass

#Principal buscar por criterio
    def bucarcriterio(self, type = None, size = None):
        result = []
        if type is None and size is not None:
            result = self.search_by_size(self.root, size)
        elif type is not None and size is None:
            result = self.search_by_type(self.root, type)
        elif type is not None and size is not None:
            result = self.search_by_type_and_size(self.root, type, size)
        return result
    
    #Buscar por tamaño
    def search_by_size(self, node, size):
        result = []
        if node:
            if node.size == size:
                result.append(node)
            result.extend(self.search_by_size(node.left, size))
            result.extend(self.search_by_size(node.right, size))
        return result
    
    #Buscar por tipo
    def search_by_type(self, node, type):
        result = []
        if node:
            if node.category == type:
                result.append(node)
            result.extend(self.search_by_type(node.left, type))
            result.extend(self.search_by_type(node.right, type))
        return result
    
    #Buscar por tipo y tamaño
    def search_by_type_and_size(self, node, type, size):
        result = []
        if node:
            if node.category == type and node.size == size:
                result.append(node)
            result.extend(self.search_by_type_and_size(node.left, type, size))
            result.extend(self.search_by_type_and_size(node.right, type, size))
        return result

#Recorrido por niveles
    def recorridoniveles(self):
        pass

#Obtener altura
    def height(self, node):
        if not node:
            return 0
        return node.height
    
#Actualizar Altura
    def update_height(self, node):
        node.height = 1 + max(self.height(node.left), self.height(node.right))

#Obtenrer el balance
    def balance(self, node):
        if not node:
            return 0
        return self.height(node.left) - self.height(node.right)
    
#Rotar a la izquierda
    def rotate_left(self,z):
        y = z.right
        T2 = y.left

        y.left = z
        z.right = T2

        self.update_height(z)
        self.update_height(y)

        return y
    
#Rotar a la derecha
    def rotate_right(self, z):
        y = z.left
        T3 = y.right

        y.right = z
        z.left = T3

        self.update_height(z)
        self.update_height(y)

        return y
```
